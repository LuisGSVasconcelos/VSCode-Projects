```python
import numpy as np
import matplotlib.pyplot as plt

A = 1
f = 2
phi = 0
t1 = np.linspace(0, 1, 1000)
x = A * np.sin(2 * np.pi * f * t1 + phi)

x1 = x

plt.plot(t1, x1)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.title('Continuous-Time Sine Wave')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time vector
t5 = np.linspace(0,1, 1000)

# Define the parameters of the Gaussian function
A = 1 # Amplitude
mu = 0 # Mean
sigma = 1 # Standard deviation

# Generate the Gaussian function
x = A * np.exp(-(t5-mu)**2 / (2*sigma**2))

x6 = 10*x

# Plot the waveform
plt.plot(t5, x6)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Gaussian Function')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constants
A = 1
B = 2
C = 3

# Generate the time axis
t8 = np.linspace(0, 1, 1000)

# Generate the parabolic signal
x = A*t8**2 + B*t8 + C

x8 = x

# Plot the signal
plt.plot(t8, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Parabolic Signal')
plt.grid()
plt.show()
```

```python
def signal_gen():

    A1 = np.random.normal(0, 1, 1)
    f1 = np.random.normal(0, 2, 1)
    phi = 0
    t = np.linspace(0, 1, 1000)
    x_sen = A1 * np.sin(2 * np.pi * f1 * t + phi)

    # Define the constants
    A2 = np.random.normal(0, 3, 1)
    B2 = np.random.normal(0, 3, 1)
    C2 = np.random.normal(0, 3, 1)

    # Generate the parabolic signal
    x_par = A2*t**2 + B2*t + C2

    x_sig = x_sen + x_par

    fs = 1000 # sampling frequency
    T = 1/fs # sampling period
    f3 = np.random.normal(0, 5, 1) # frequency of square wave

    x_square = np.sign(np.sin(2*np.pi*f3*t)) # generate square wave

    x_rand = np.random.normal(0, 0.5, len(t))*np.random.normal(0, 1, len(t))

    x_sig = x_sen + x_par + x_square + x_rand

    return x_sig

x_sig = signal_gen()

plt.plot(t, x_sig)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Signal')
plt.show()
```
