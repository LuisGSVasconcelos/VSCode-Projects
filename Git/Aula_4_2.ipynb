```python
# read noisy flow data 
import numpy as np 
noisy_signal = np.loadtxt('noisy_flow_signal.csv', delimiter=',') 
```

```python
noisy_signal
```

```python
import matplotlib.pyplot as plt
plt.plot(noisy_signal)
```

```python
# denoise using SMA filter 
import pandas as pd 
windowSize = 15 
smoothed_signal_MA = pd.DataFrame(noisy_signal).rolling(windowSize).mean().values # taking mean on a rolling window and converting back to numpy array
```

```python
plt.plot(noisy_signal, label= 'noisy_signal')
plt.plot(smoothed_signal_MA, label= 'smoothed_signal_MA')
plt.legend()
```

```python
# import savgol_filter from scipy 
from scipy.signal import savgol_filter 
smoothed_signal_SG = savgol_filter(noisy_signal, window_length = 15, polyorder = 3)
```

```python
plt.plot(noisy_signal, label= 'noisy_signal')
plt.plot(smoothed_signal_MA, label= 'smoothed_signal_MA')
plt.plot(smoothed_signal_SG, 'y-' ,label= 'smoothed_signal_SG')
plt.legend()
```
